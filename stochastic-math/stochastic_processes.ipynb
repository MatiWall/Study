# Stochastic Process
A stochastic process is a sequence of random variables in a probability space where the index of the sequence usually has the interpretation of time.

There are both discrete and continuous processes. the key differences are:

1. Discrete
    1. Indexing parameter is discrete (countable). i.e. you known where to start counting (positive natural numbers vs positive real number)
    2. Each index is associated with a discrete random variable usually taking a finite or countable infinite set.
    3. Examples could be a coin flip or the outcome of a dice roll.
2. Continues
    1. Indexing parameter is continuous (non countable), usually could be time.
    2. At each index point the process is associated with random variables that can take on values from a continuous range.

**Time Granularity**: Discrete processes evolve in steps or intervals of time or space, while continuous processes evolve continuously over a range of values. In discrete processes, time is usually divided into fixed increments, while continuous processes operate with infinitesimal increments.

Examples of stochastic process are

**Wiener Process**: The Wiener process is a stochastic process with stationary and independent increments that are normally distributed based on the size of the increments (See notebook wiener_process)

**Random Walk**: 
Random walks are stochastic processes that are usually defined as sums of i.i.d random variables or random vectors in Euclidean space, so they are processes that change in discrete time.

**Bernoulli Process**
In probability and statistics, a Bernoulli process (named after Jacob Bernoulli) is a finite or infinite sequence of binary random variables, so it is a discrete-time stochastic process that takes only two values, canonically 0 and 1.


Depending on their mathematical properties these random processes can be grouped into various categories such as:
* Random Walks: Sequences of steps with random increments, modeling various random phenomena.
* Martingales: Processes where future expectations are equal to current values, often used in fair game theory.
* Markov Processes: Memoryless processes where future states depend only on the current state.
* Levy Processes: Processes with stationary, independent increments, including both continuous and jump components.
* Gaussian Processes: Processes where any finite collection of variables follows a joint Gaussian distribution, used in function modeling.
* Random Fields: Multi-dimensional generalizations of random processes for spatial or temporal variations.

## Definition

A stochastic process is defined as the collections of random variables on a common probability space $(\Omega, \mathcal{F}, P)$ (also known as a probability triplet) and the random variable, indexed by some set $T$, all takes values in the same space $S$ which has to be measurable with respect to some $\sigma$- algebra $\Sigma$.  In another way for a given probability space $(\Omega, \mathcal{F}, P)$ and measurable space $(S, \Sigma)$ is a collection of $S$ valued random variables written as $\{X(t): t\in T\}$.

> $\sigma$-algebra: A $\sigma$-algebra on a set $S$ is a nonempty collection $\Sigma$ of subsets of $S$ closed under complement, countable unions and countable intersections.

The sample space represents the collection of all possible outcomes of a stochastic process, while the measurable space defines which subsets of the sample space are considered "measurable" and can have probabilities assigned to them.

> **Difference between $\Omega$ and $S$** 
> $\Omega$ is the space of all outcomes of the underlying random variable i.e. the elements of $\Omega$ are the basic outcomes that can occur. $S$ is the space in which the random variable associated with the process takes values. As an example consider a roll with a six-sided die. The sample space is $\{1,2,3,4,5,6\}$. Now let $X$ be a random variable representing the outcome of the die roll plus 10. In this case $S=\{11,12,13,14,15,16\}$ i.e. $\Omega$ and $S$ are different. A random variable is a stochastic process that map outcomes from $\Omega$ to $S$.

The probability space consists of three elements:
1. The sample space $\Omega$, i.e. the set of all possibilities.
2. An event space $\mathcal{F}$, where an event is a set of outcomes in he sample space (subsets of  $\Omega$).
3. A probability function $P$ assigning each event with a probability (probability measure).

Considering a standard die throw example with sample space is $\{1,2,3,4,5,6\}$. We can choose the event space as the set of all subsets. This contains simple events such as $\{2\}$ i.e. the die lands on 2 and $\{2,4,6\}$ i.e. the die lands on an even number. The probability function then maps each of these event to a probability i.e. the number of outcomes in an event divided by the total number of outcomes int the sample space. For the event $\{2\}$ this is trivial as there is only a single outcome in the event space i.e. $P(\{2\}) = 1/6$, whereas for $\{2,4,6\}$ there is three outcomes i.e. $P(\{2,4,6\})=3/6=1/2$. Considering an experiment, we imagine that nature select a single outcome $\omega$ from the sample space $\Omega$. An event in the event space $\mathcal{F}$ that contains $\omega$ are said to have occurred. By repeating the experiment many time the probability for each event should match that assigned by $P$.

### σ-Algebra (Sigma-Algebra)

**Definition**: A $\sigma$-algebra $\mathcal{F}$ over a sample space $\Omega$ is a collection of subsets of $\Omega$ that satisfies the following properties:
  1. **Contains the Sample Space**: $\Omega \in \mathcal{F}$
  2. **Closed Under Complements**: If $A \in \mathcal{F}$, then $A^c = \Omega \setminus A \in \mathcal{F}$
  3. **Closed Under Countable Unions**: If $A_1, A_2, A_3, \ldots \in \mathcal{F}$, then $\bigcup_{i=1}^\infty A_i \in \mathcal{F}$

**Definition**: For a collection of subsets $\{A_i\}$ of $\Omega$, the σ-algebra $\sigma(\{A_i\})$ is the smallest σ-algebra containing every set $A_i$ from the collection. It includes all possible events that can be described using the sets in $\{A_i\}$ through countable unions, intersections, and complements.

Let’s consider the following example:

**Sample Space**: $\Omega = \{1, 2, 3, 4, 5, 6\}$

**Sets Given**: $\{1\}$ and $\{4\}$

**Generating σ-Algebra**: $\sigma\{\{1\}, \{4\}\}$ is the smallest $\sigma$-algebra that contains the sets $\{1\}$ and $\{4\}$.

To find $\sigma\{\{1\}, \{4\}\}$, we need to consider all possible sets formed from unions, intersections, and complements of $\{1\}$ and $\{4\}$:

1. **Original Sets**:
   - $\{1\}$
   - $\{4\}$

2. **Complements**:
   - $\{1\}^c = \{2, 3, 4, 5, 6\}$
   - $\{4\}^c = \{1, 2, 3, 5, 6\}$

3. **Unions**:
   - $\{1\} \cup \{4\} = \{1, 4\}$
   - $\{1\}^c \cup \{4\}^c = \{2, 3, 5, 6\}$

4. **Intersections**:
   - $\{1\} \cap \{4\} = \emptyset$
   - $\{1\} \cap \{4\}^c = \{1\}$
   - $\{4\} \cap \{1\}^c = \{4\}$

5. **The Entire Sample Space**: $\Omega = \{1, 2, 3, 4, 5, 6\}$

Thus, $\sigma\{\{1\}, \{4\}\}$ includes:
- $\emptyset$
- $\{1\}$
- $\{4\}$
- $\{1\}^c = \{2, 3, 4, 5, 6\}$
- $\{4\}^c = \{1, 2, 3, 5, 6\}$
- $\{1, 4\} = \{1, 4\}$
- $\{1, 4\}^c = \{2, 3, 5, 6\}$
- $\Omega = \{1, 2, 3, 4, 5, 6\}$


The $\sigma$-algebra $\sigma\{\{1\}, \{4\}\}$ contains all the events that can be determined from knowing whether a 1 or a 4 has occurred or not. This means you can describe all possible scenarios involving these sets, including their intersections, unions, and complements, which fully characterize the possible outcomes and their probabilities.

This σ-algebra encapsulates the information we can extract from the observations of whether or not the specific outcomes (e.g., a 1 or a 4) have occurred.


## Probability Space
A probability measure is a fundamental concept in probability theory and measure theory that provides a way to assign probabilities to events within a sample space. It formalizes the intuitive notion of probability and satisfies specific axioms to ensure its consistency and usefulness. Here's an explanation of a probability measure:
1. Non-negativity: $P(A_i) \ge 0$ for $A_i \in \mathcal{F}$
2. Normalization: $P(\Omega) = 1$
3. Countable additive: $P(A_1 \cup A_2 \cup A_3 \cup ...) = P(A_1)+P(A_2)+P(A_3)+...$ 

The probability measure - a function on $\mathcal{F}$such that $P$ is countable additive i.e. a function mapping sets to numbers, with the property that its value on a union of two disjoint sets equal the sum of its values namely $\{A_i\}_{i=0}^{\infty} \subseteq \mathcal{F}$ then $P\left( \cup_{i=1}^{\infty} A_i \right) = \Sigma_{i=1}^{\infty}P(A_i)$. In addition the measure of the entire sample space is equal to one i.e. $P(\Omega) = 1$.
